# **Stack implementation using Arrays**

***While Python has built-in lists, I'm implementing a stack manually using arrays to better understand the logic behind it. This hands-on approach clarifies how stacks work, including operations like push, pop, and traverse, as well as concepts like overflow and underflow.***

```python
class stack:

  def __init__(self, size):
    self.size = size
    self.array = [None] * self.size
    self.top = -1


  def push(self, value):
    if self.top == self.size - 1:
      return 'Overflow'
    else:
      self.top += 1
      self.array[self.top] = value


  def pop(self):
    if self.top == -1:
      return 'Underflow - Empty stack'
    else:
      data = self.array[self.top]
      self.top -= 1

      print(data)


  def traverse(self):
    for i in range(self.top + 1):
      print(self.array[i], end=' ')
```

```python
s = stack(3)
```

```python
s.push(9)
s.push(8)
s.push(1)
```

```python
s.push(4)
```

```python
s.pop()
```

```python
s.traverse()
```
