# **Queue Implemetation using LL**

*In this implementation of a **queue** using a **linked list**, I perform several key operations. First, I **enqueue** elements by adding them to the rear of the linked list. Then, I can **dequeue** elements by removing them from the front. The **size** function allows me to check the number of elements in the queue, while the **isEmpty** function verifies if the queue is empty. Lastly, I use the **traverse** function to display all the elements in the queue from front to rear, ensuring smooth navigation through the linked list.*

```python
class Node:

  def __init__(self,value):
    self.data = value
    self.next = None
```

```python
class Queue:

  def __init__(self):
    self.front = None
    self.rear = None


  def enqueue(self, value):

    newnode = Node(value)

    if self.rear == None:
      self.front = newnode
      self.rear = self.front
    else:
      self.rear.next = newnode
      self.rear = newnode


  def dequeue(self):

    if self.front == None:
      return 'Empty'
    else:
      self.front = self.front.next


  def is_empty(self):
    return self.front == None


  def size(self):

    temp = self.front
    count = 0

    while temp != None:
      count += 1
      temp = temp.next

    return count


  def front_value(self):
    if self.front == None:
      return "Empty Queue"
    else:
      return self.front.data


  def rear_value(self):

    if self.front == None:
      return 'Empty Queue'
    else:
      return self.rear.data


  def traverse(self):

    temp = self.front

    while temp != None:
      print(temp.data, end=' ')
      temp = temp.next

```

```python
q = Queue()
```

```python
q.enqueue(7)
q.enqueue(8)
q.enqueue(9)
q.enqueue(5)
```

```python
q.traverse()
```

```python
q.dequeue()
```
