```python
class Node:
  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
class LinkedList:
  def __init__(self):

    self.head = None  # empty list
    self.n = 0  # count of nodes

  def __len__(self):
    return self.n


  def insert_head(self, value):

    new_node = Node(value)  #newnode
    new_node.next = self.head
    self.head = new_node

    self.n += 1


  def insert_last(self, value):     # append

    new_node = Node(value)

    if self.head == None:
      self.head = new_node
      self.n += 1
      return

    temp = self.head

    while temp.next != None:    # travelling to the last node
      temp = temp.next

    temp.next = new_node
    self.n += 1


  def insert_random(self, middle_data, value):

    new_node = Node(value)

    if self.head == None:     # if list is empty
      new_node = self.head
      self.n += 1
      return

    temp = self.head
    while temp != None:
      if temp.data == middle_data:    # checking of the middle data
        break
      temp = temp.next

    if temp.data == middle_data:
      new_node.next = temp.next
      temp.next = new_node
      self.n += 1
    else:
      return 'Item Not Found!'


  def traverse(self):     # print

    if self.head == None:
      return 'Empty Linked List'

    temp = self.head
    result = ''

    while temp != None:
      result += str(temp.data) + '->'
      temp = temp.next

    return result[:-2]


  def clear(self):
    self.head = None
    self.n = 0


  def delete_head(self):

    if self.head == None:
      return 'Empty Linked List'

    self.head = self.head.next
    self.n -= 1


  def delete_last(self):    # pop

    if self.head == None:
      return 'Empty Linked List'

    temp = self.head

    if temp.next == None:     # If only one node is in the list
      self.head = None
      self.n -= 1
      return

    while temp.next != None:      # another eg ( while temp.next.next != None )
      prev = temp
      temp = temp.next

    prev.next = None
    self.n -= 1


  def delete_value(self, value):      # remove

    if self.head == None:
      return 'Empty Linked List'

    if self.head.data == value:
      return self.delete_head()

    temp = self.head

    while temp != None:
      if temp.data == value:
        break
      prev = temp
      temp = temp.next

    if temp == None:
      return 'Item Not Found'
    else:
      prev.next = temp.next
      self.n -=1


  def search(self, value):

    # if self.head == None:
    #   return 'Empty Linked List'

    temp = self.head
    loc = 0

    while temp != None:
      if temp.data == value:
        return loc
      temp = temp.next
      loc += 1

    return 'Item Not Found'


  def __getitem__(self, index):

    temp = self.head
    loc = 0

    while temp != None:
      if loc == index:
        return temp.data
      temp = temp.next
      loc += 1

    return 'IndexError'


  def replace_max(self, value):   # replacing the number with the highest number

    if self.head == None:
      return 'Empty Linked List'

    temp = self.head
    max = temp

    while temp != None:
      if temp.data > max.data:
        max = temp
      temp = temp.next

    max.data = value


  def sum_odd_nodes(self):    # index odd number

    if self.head == None:
      return 'Empty Linked List'

    temp = self.head
    counter = 0
    result = 0

    while temp != None:
      if counter %2 != 0:   # checking if it's odd index
        result = result + temp.data
      counter += 1
      temp = temp.next

    print(result)


  def reversal(self):

    if self.head == None:
      return 'Empty Linked List'

    prev = None     # initially none because the last node.next is always none
    temp = self.head

    while temp != None:
      next = temp.next
      temp.next = prev
      prev = temp
      temp = next

    self.head = prev  # anaylze the while loop

```
