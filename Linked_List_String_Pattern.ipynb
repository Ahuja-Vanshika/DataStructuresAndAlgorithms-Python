# **Linked List String Problem (With Special Characters):**

Given a singly linked list where each node contains a single character, the task is to traverse the linked list and form a string with specific formatting rules:

*   If two consecutive special characters like /*, **, or // are encountered, the next character should be converted to uppercase and followed by a space.

*   If a single special character is encountered, only a space should be added after it.

**Example:**

For a linked list like [H -> i -> / -> * -> p -> i -> n -> k -> None], the resulting string would be "Hi Pink".



```python
class Node:
  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
class LinkedList:
  def __init__(self):

    self.head = None  # empty list


  def append(self, value):

    new_node = Node(value)

    if self.head == None:
      self.head = new_node
      return

    temp = self.head

    while temp.next != None:    # travelling to the last node
      temp = temp.next

    temp.next = new_node


  def traverse(self):     # print

    if self.head == None:
      return 'Empty Linked List'

    temp = self.head
    result = ''

    while temp != None:
      result += str(temp.data) + '-> '
      temp = temp.next

    return result[:-2]


  def create_sentence(self):

    temp = self.head

    while temp != None:
      if temp.data == '*' or temp.data == '/':
        temp.data = ' '

        if temp.next.data == '*' or temp.next.data == '/':
          temp.next.next.data = temp.next.next.data.upper()     # to skip the special characters
          temp.next = temp.next.next    # to skip over the upper cased letter

      temp = temp.next


  def traversal(self):      # here just to remove the arrows
    if self.head == None:
      return 'Empty Linked List'

    temp = self.head
    result = ''

    while temp != None:
      result += str(temp.data)
      temp = temp.next

    return result
```
